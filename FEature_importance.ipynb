#### Discrete Variables  
A variable which values are whole numbers (counts) is called  
discrete. Examples:  
 Number of items bought by a customer in a supermarket (10, 50, )  
 Number of active bank accounts of a borrower (1, 4, 7, ...)  
 Number of pets in the family  
 Number of children in the family

### Continuous Variables  
A variable that may contain any value within some range is called  
continuous. Examples:  
 Amount paid by a customer in a supermarket ($32.50, $12, $5.20, )  
 House price (GBP 350,000, GBP 57000, GBP 1,000,000, ...)  
 Time spent surfing a website (3.4 seconds, 5.10 seconds, ...)  
 Total debt as percentage of total income in the last month (0.2, 0.001, 0,  
0.75, ...)


### Ordinal Variables  
Categorical variables in which categories can be meaningfully  
ordered are called ordinal. Examples:  
 Student's grade in an exam (A, B, C or Fail)  
 Days of the week (Monday = 1 and Sunday = 7)  
 Educational level, with the categories: Elementary school, High  
school, College graduate and PhD ranked from 1 to 4

### Nominal Variables  
Show no intrinsic order of the labels. Examples:  
 Country of birth (Argentina, England, Germany)  
 Postcode  
 Vehicle make (Citroen, Peugeot, ...)

## what is the differnece between Pearson's r and Spearman's

The main difference between them is 
	
 - the type of data they can be used with and the 
 - nature of the relationship they measure.

| Pearson's  | Spearman's |
|--|--|
|measure the **linear** correlation between two **continuous** variables  | measure the **monotonic** correlation between two variables	 |
| It ranges from -1 (a perfect negative correlation) to 1 (a perfect positive correlation), with 0 indicating no correlation. | it measures the degree to which the two variables are related, regardless of whether the relationship is linear or nonlinear. Spearman's rho ranges from -1 to 1, with 0 indicating no correlation	 |
| assumes that the variables are normally distributed and that the relationship between them is linear. | **does not assume** that the variables are **normally distributed** and can be used with any type of data, including ordinal data.	 |
|  | 	 |

**Pearson's r:**

1.  Calculate the mean of x (x) and y (y).
    
2.  Calculate the standard deviation of x (s_x) and y (s_y).
    
3.  Calculate the covariance between x and y (cov_xy).
    
4.  Calculate Pearson's correlation coefficient (r) using the following formula:
    
    r = cov_xy / (s_x * s_y)

**Spearman's rho:**

1.  Calculate the rank of each value in x and y.
    
    -   Assign each value in x and y a rank, starting with 1 for the smallest value and increasing by 1 for each subsequent value. Ties are assigned the average of their ranks.
2.  Calculate the difference between the ranks of each pair of corresponding values (d).
    
3.  Calculate the sum of the squared differences (d).
    
4.  Calculate Spearman's rank correlation coefficient () using the following formula:
    
     = 1 - ((6 * d) / (n * (n - 1)))



### What is Feature Importance?

 - Feature Importance refers to techniques that calculate a score for
   all the input features for a given model.  
 - A higher score means that the specific feature will have a larger effect on the model that is being used to predict a certain variable.

#### Permutation Feature Importance

> The feature importance is calculated by noticing the increase or decrease in error when we permute the values of a feature
> If permuting the values causes a huge change in the error, it means the feature is important for our model
> its model agnostic approch and can be applied to every machine learning model. 

#### The permutation feature importance is based on an algorithm that works as 
follows.

1.  Calculate the **mean squared error with the original values**
2.  Shuffle the values for the features and make predictions
3.  Calculate the mean squared error with the shuffled values
4.  Compare the difference between them
5.  Sort the differences in descending order to get features with most to least importance
 or
 
6.  Train a baseline model and record the score (we use accuracy in this example) on the validation set.
7.  Re-shuffle values for one feature, use the model to predict again, and calculate scores on the validation set. The feature importance for the feature is the difference between the baseline in 1 and the permutation score in 2.
8.  Repeat the process for all features.

**Pearson's Correlation:** Pearson's correlation measures the linear correlation between two variables. `It ranges from -1 to +1, where -1` indicates a strong negative correlation,` 0 indicates no correlation`, and `+1 indicates a strong positive correlation`.

> Pearson's correlation can be used to select features that are highly correlated with the target variable.

**Spearman's Correlation**: Spearman's correlation measures the monotonic correlation between two variables. It ranges from 
- -1 to +1, where -1 indicates a strong negative monotonic correlation, 
- 0 indicates no monotonic correlation, and 
- +1 indicates a strong positive monotonic correlation.

> Spearman's correlation can be used to select features that are highly correlated with the target variable, regardless of the linearity of the relationship.

> Winsorization is a data preprocessing technique that involves `limiting the extreme values of a feature` to reduce their impact on the model. 
> Specifically, it involves replacing the `extreme values with the next largest or smallest values that are within a specified percentile range`

Pros.
- dealing with outliers that can significantly affect the model's performance.
- limiting the impact of extreme values
- can improve the stability and robustness of the model.

Cons:
- It can result in the loss of information and 
- can affect the distribution and variability of the data
