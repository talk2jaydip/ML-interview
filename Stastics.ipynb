#### binomial distribution
The binomial distribution is a probability distribution `that describes the number of successes in a fixed number of independent trials`, where each trial has only two possible outcomes (success or failure), and the probability of success is the same for each trial.

For example, let's say we want to know the probability of flipping a coin and getting heads 3 times in a row out of 5 total flips. We can model this situation using the binomial distribution, where:

-   The number of trials (n) is 5, since we're flipping the coin 5 times.
-   The probability of success (p) is 0.5, since the probability of getting heads on a fair coin is 0.5.
-   The number of successes we want (k) is 3, since we want to get heads 3 times in a row.

Using the binomial distribution formula, we can calculate the probability of getting exactly 3 heads in 5 flips as:

P(X = 3) = (5 choose 3) * (0.5)^3 * (1 - 0.5)^(5-3) = 0.3125

where (5 choose 3) is the number of ways to choose 3 flips out of 5 to be heads, (0.5)^3 is the probability of getting 3 heads, and (1 - 0.5)^(5-3) is the probability of getting 2 tails.

So the probability of getting exactly 3 heads in 5 flips of a fair coin is 0.3125, or about 31.25%.

What does it mean to have high and low p-values?
- **A high p-value** (close to 1) suggests that the observed results are likely to have occurred by chance, and there is not enough evidence to reject the null hypothesis.
- **A low p-value** (close to 0) suggests that the observed results are unlikely to have occurred by chance, and there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

> A high p-value means that the observed result is consistent with the null hypothesis, and there is not enough evidence to reject it. For example, if we want to test whether a coin is fair, and we get a p-value of 0.8 after tossing it 10 times and getting 6 heads and 4 tails, we would say that this result is likely to happen by chance even if the coin is fair, and we would not reject the null hypothesis that the coin is fair.

> A low p-value means that the observed result is inconsistent with the null hypothesis, and there is enough evidence to reject it. For example, if we want to test whether a drug has an effect on a disease, and we get a p-value of 0.01 after giving it to 100 patients and observing that 80 of them recover, we would say that this result is unlikely to happen by chance if the drug has no effect, and we would reject the null hypothesis that the drug has no effect.
