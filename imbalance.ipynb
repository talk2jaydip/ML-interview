#### ADASYN  
 Uses a weighted distribution of the minority class  according to how difficult the observations are to be learned or classified.

**Oversampling:**

Pros:

-   Increases the representation of the `minority class`, which can improve the performance of machine learning models on imbalanced datasets.
-   Does not discard any data points, which can be important `if the dataset is already small`.

Cons:

-   `Can lead to overfitting` if the same data points are used in both the training and testing sets.
-   `Can increase the risk of data leakage` if the same data points are used in both the training and testing sets.

**Undersampling:**

Pros:

-   `Reduces the amount of data in the dataset`, which can make it easier and faster to train machine learning models.
-   Helps to `balance the classes` in the dataset, which can improve the performance of machine learning models on imbalanced datasets.

Cons:

-   Discards data points, which can lead to `loss of information` and reduce the representativeness of the dataset.
-   `Can lead to bias` if important information is lost due to undersampling, or if the undersampling technique used is not appropriate for the dataset.

#### QnA


18.  What are Tomek links?

Answer: Tomek links are pairs of instances from different classes that are closest to each other, and removing the instances of the majority class that form Tomek links can help to balance an imbalanced dataset.

19.  What is class weighting?

Answer: Class weighting is a technique that assigns higher weights to the minority class and lower weights to the majority class to balance the dataset.

20.  What is cost-sensitive learning?

Answer: Cost-sensitive learning is a technique that takes into account the cost of misclassification for different classes.

5.  Can you describe a scenario where ensemble learning can be useful for developing a model on an imbalanced dataset?

Answer: Ensemble learning can be useful when there is a high degree of class imbalance and it is difficult to obtain a balanced dataset using undersampling or oversampling techniques.

2.  Can you explain the concept of threshold tuning in the context of imbalanced datasets?

Answer: Threshold tuning is the process of adjusting the probability threshold of a model to optimize its performance for a specific evaluation metric, such as F1 score or AUC-ROC, on an imbalanced dataset. This is important because the default threshold of 0.5 may not be the optimal threshold for imbalanced datasets.
